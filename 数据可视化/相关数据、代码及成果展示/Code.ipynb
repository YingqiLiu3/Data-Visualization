```python
import pandas as pd
df = pd.read_excel('各级各类学历教育毕业生数.xlsx')
df
```

```python
import matplotlib.pyplot as plt

plt.rcParams['font.family'] = ['SimHei']  # 添加这条可以让图形显示中文


plt.figure(figsize=(10,6))
# plot中参数的含义分别是横轴值，纵轴值，线的形状，颜色，透明度,线的宽度和标签
#plt.plot(df.columns[2:], df.iloc[0][2:].values,  'o-',alpha=0.8, linewidth=1, label=df["指标"][0])
plt.plot(df.columns[1:], df.iloc[1][1:].values,  'o-',alpha=0.8, linewidth=1, label=df["指标"][1])
plt.plot(df.columns[1:], df.iloc[2][1:].values,  'o-',alpha=0.8, linewidth=1, label=df["指标"][2])
plt.plot(df.columns[1:], df.iloc[4][1:].values,  'o-',alpha=0.8, linewidth=1, label=df["指标"][4])
plt.plot(df.columns[1:], df.iloc[5][1:].values,  'o-',alpha=0.8, linewidth=1, label=df["指标"][5])
plt.plot(df.columns[1:], df.iloc[7][1:].values,  'o-',alpha=0.8, linewidth=1, label=df["指标"][7])
plt.plot(df.columns[1:], df.iloc[8][1:].values,  'o-',alpha=0.8, linewidth=1, label=df["指标"][8])
# 显示标签，如果不加这句，即使在plot中加了label='一些数字'的参数，最终还是不会显示标签
plt.legend(fontsize=10)
plt.xlabel('年份',fontsize=15)
plt.ylabel('人数',fontsize=15)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.title('各级各类学历教育毕业生数')
plt.grid()
plt.show()
# plt.savefig('demo.jpg')  # 保存该图片
```

```python
import matplotlib.pyplot as plt
nums=[1,2,4,5,7,8]

df1=pd.DataFrame(df,index=nums)
df1
```

```python
plt.figure(figsize=(5,5))
patches,l_text,p_text=plt.pie(df1["2020年"], labels=df1["指标"],autopct='%.2f%%')
for t in l_text:
    t.set_size(15)
for t in p_text:
    t.set_size(15)
plt.title("2022年各级各类学历教育毕业生比例",fontsize=15)
plt.show()
```

```python
plt.figure(figsize=(5,5))
patches,l_text,p_text=plt.pie(df1["2019年"], labels=df1["指标"],autopct='%.2f%%')
for t in l_text:
    t.set_size(15)
for t in p_text:
    t.set_size(15)
plt.title("2021年各级各类学历教育毕业生比例",fontsize=15)
plt.show()
```

```python
#获取每年的百分比
data=df1.iloc[:,1:].div(df1.iloc[:,1:].sum(axis=0), axis=1)
data
```

```python
plt.figure(figsize=(10,5))
for i in range(0,6):
    plt.plot(data.columns, data.iloc[i,:].values, 'o-',alpha=0.8, linewidth=1, label=df1["指标"].iloc[i])
plt.xlabel('年份',fontsize=15)
plt.ylabel('占比',fontsize=15)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.title("各级各类学历教育毕业生占比图")
plt.legend()
plt.grid()
plt.show()
```

```python
import pandas as pd
data = pd.read_excel('普通高等学校数（所）.xlsx') 
data
```

```python

# 需要引用的库
from pyecharts import options as opts
from pyecharts.charts import Map
import random

school = data.iloc[:,:].values
def create_china_map():
    ''' 
     作用：生成中国地图
    '''
    (
        Map()
        .add(
            series_name="学校数", 
            data_pair=school, 
            maptype="china", 
        )
        .set_global_opts(
        # 设置标题
        title_opts=opts.TitleOpts(title="普通高等学校数(所)"),
        # 设置标准显示
        visualmap_opts=opts.VisualMapOpts(max_=168, is_piecewise=False)
        )
        # 生成本地html文件
        .render("普通高等学校数(所).html")
    )

create_china_map()
```
